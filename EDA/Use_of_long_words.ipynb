# Import the python libraries

```python
# importing libraries being used in this notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read in the dataset

```python
# reading the dataset for fake news
news_df = pd.read_csv('./Datasets/fake-news/train.csv',index_col=0)
# drop NaN values in text column from news_df dataset
news_df.dropna(subset=['text'], inplace=True)
news_df.head()
```

# Determine if the use of long words is a good predictor of fake/real news


```python
# This code will determine  if the eu of long words [more than n characters long] is a good predictor of fake news
# The code will also determine the optimal value of n

# creating a new column in the dataframe to store the number of words in the text
news_df['num_words'] = news_df['text'].apply(lambda x: len(x.split()))

# creating a new column in the dataframe to store the number of words in the text that are more than 6 characters long
news_df['num_long_words'] = news_df['text'].apply(lambda x: len([w for w in x.split() if len(w)>6]))

# creating a new column in the dataframe to store the ratio of long words to total words in the text
news_df['ratio_long_words'] = news_df['num_long_words']/news_df['num_words']

# creating a new column in the dataframe to store the number of words in the text that are more than 10 characters long
news_df['num_very_long_words'] = news_df['text'].apply(lambda x: len([w for w in x.split() if len(w)>10]))

# creating a new column in the dataframe to store the ratio of very long words to total words in the text
news_df['ratio_very_long_words'] = news_df['num_very_long_words']/news_df['num_words']

# creating a new column in the dataframe to store the number of words in the text that are more than 15 characters long
news_df['num_very_very_long_words'] = news_df['text'].apply(lambda x: len([w for w in x.split() if len(w)>15]))

# creating a new column in the dataframe to store the ratio of very very long words to total words in the text
news_df['ratio_very_very_long_words'] = news_df['num_very_very_long_words']/news_df['num_words']

# creating a new column in the dataframe to store the number of words in the text that are more than 20 characters long
news_df['num_very_very_very_long_words'] = news_df['text'].apply(lambda x: len([w for w in x.split() if len(w)>20]))

# creating a new column in the dataframe to store the ratio of very very very long words to total words in the text
news_df['ratio_very_very_very_long_words'] = news_df['num_very_very_very_long_words']/news_df['num_words']

# creating a new column in the dataframe to store the number of words in the text that are more than 25 characters long
news_df['num_very_very_very_very_long_words'] = news_df['text'].apply(lambda x: len([w for w in x.split() if len(w)>25]))

# creating a new column in the dataframe to store the ratio of very very very very long words to total words in the text
news_df['ratio_very_very_very_very_long_words'] = news_df['num_very_very_very_very_long_words']/news_df['num_words']

```

```python
# Plotting the histograms of the ratio of long words to total words in the text for fake and real news
plt.figure(figsize=(12,6))
plt.hist(news_df[news_df['label']==0]['ratio_long_words'], bins=20, alpha=0.7, label='Real')
plt.hist(news_df[news_df['label']==1]['ratio_long_words'], bins=20, alpha=0.5, label='Fake')
plt.title('Histogram of ratio of long words to total words in the text')
plt.xlabel('Ratio of long words to total words in the text')
plt.ylabel('Number of news articles')
plt.legend()
plt.show()
```

```python
# Plotting the histograms of the ratio of very long words to total words in the text for fake and real news
plt.figure(figsize=(12,6))
plt.hist(news_df[news_df['label']==0]['ratio_very_long_words'], bins=20, alpha=0.7, label='Real')
plt.hist(news_df[news_df['label']==1]['ratio_very_long_words'], bins=20,  alpha=0.5, label='Fake')
plt.title('Histogram of ratio of very long words to total words in the text')
plt.xlabel('Ratio of very long words to total words in the text')
plt.ylabel('Number of news articles')
plt.legend()
plt.show()
```

```python
# Plotting the histograms of the ratio of very very long words to total words in the text for fake and real news
plt.figure(figsize=(12,6))
plt.hist(news_df[news_df['label']==0]['ratio_very_very_long_words'], bins=20, alpha=0.7, label='Real')
plt.hist(news_df[news_df['label']==1]['ratio_very_very_long_words'], bins=20,  alpha=0.5, label='Fake')
plt.title('Histogram of ratio of very very long words to total words in the text')
plt.xlabel('Ratio of very very long words to total words in the text')
plt.ylabel('Number of news articles')
plt.legend()
plt.show()
```

```python
# Plotting the histograms of the ratio of very very very long words to total words in the text for fake and real news
plt.figure(figsize=(12,6))
plt.hist(news_df[news_df['label']==0]['ratio_very_very_very_long_words'], bins=20, alpha=0.7, label='Real')
plt.hist(news_df[news_df['label']==1]['ratio_very_very_very_long_words'], bins=20,  alpha=0.5, label='Fake')
plt.title('Histogram of ratio of very very very long words to total words in the text')
plt.xlabel('Ratio of very very very long words to total words in the text')
plt.ylabel('Number of news articles')
plt.legend()
plt.show()
```
