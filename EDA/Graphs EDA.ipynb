```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Grammar

```python
df=pd.read_csv('../datasets/fake-news/train_with_grammar.csv')
```

```python
df=df[~df['text'].isna()]
df['word_count'] = df['text'].apply(lambda x: len(x.split()))
df=df[df['word_count']>0]
```

```python
df['gram_errors_rel']=df['Grammar Errors']/df['word_count']
```

```python
plt.figure(figsize=(8, 6))

# Plot the KDE for dataframe 1
sns.kdeplot(data=df[df['label']==1]['gram_errors_rel'], label='Fake', shade=True)

# Plot the KDE for dataframe 2
sns.kdeplot(data=df[df['label']==0]['gram_errors_rel'], label='True', shade=True)

# Set the plot title and labels
plt.title(f'Kernel Density Estimation for Grammatical Errors (rel to N of Words)')
plt.xlabel('Gram Errors prop')
plt.ylabel('Density')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
print([df[df['label']==1]['gram_errors_rel'].dropna().mean(),df[df['label']==0]['gram_errors_rel'].dropna().mean()])
```
